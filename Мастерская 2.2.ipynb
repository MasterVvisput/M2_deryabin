### Функция `get_df_info` 
Принимает на вход датафрейм данных, возвращает датафрейм с: 
- количеством строк
- количеством признаков
- количеством пропусков
- количеством дублей

````python
def get_df_info(data):
  num_cols = data.select_dtypes(include=['float64','int64'])
  summ = pd.DataFrame(num_cols.dtypes, columns=['data type'])
  rows = data.shape[0]
  features = data.shape[1]
  isna_count_sum = data.isna().sum().sum()
  duplicateds = data.duplicated().sum().sum()
  res = pd.DataFrame({'rows': rows, 'features': features,
                      'isna_count_sum': isna_count_sum,
                      'duplicates': duplicateds}, index = [0])
  return res

hide_toggle()
````

### Функция расчета и вывода процентов выбросов для выбранных признаков
- `col: pd.Series` это основной аргумент функции, предполагает на вход серию данных pandas
- `whiskers=False` опциональный аргумент, который определяет, должна ли функция возвращать "усы" - границы допустимых значений. 
- `iqr_coeff=1.5` опциональный аргумент, который определяет коэффециент, используемый при расчете межквартильного размаха.
В основе функции используется понятие межквартильнорго размаха и концепция "усов" из "ящика с усами". 

Функция проводит проверку набора данных на наличие выбросов и возвращает их долю. 


````python
# Введём функцию рассчёта и вывода процентов выбросов для отобранных
# признаков.
def get_outliers_ratio(col: pd.Series, whiskers=False, iqr_coeff=1.5):
    # Вычислим интерквартильный размах.
    distr_iqr = col.quantile(.75) - col.quantile(.25)
    # Вычислим границы.
    whisker_upp = col.quantile(.75) + iqr_coeff * distr_iqr
    whisker_low = col.quantile(.25) - iqr_coeff * distr_iqr
    # Если усы границы экстремумы, приравняем их.
    if whisker_upp >= col.max():
        whisker_upp = col.max()
    if whisker_low <= col.min():
        whisker_low = col.min()
    else:
        pass
    # Вычислим процент выбросов.
    outliers_ratio = (
        (col[col > whisker_upp].count() +
         col[col < whisker_low].count()) /
        len(col)
        )
    # Выведем границы и процент выбросов если необходимо.
    if whiskers:
        return whisker_low, whisker_upp, outliers_ratio
    # Выведем только процент выбросов.
    else:
        return outliers_ratio
    
hide_toggle()
````

### Функция построения графиков

Эта функция предназначена для визуализации исследовательского анализа данных (EDA) для заданного DataFrame. Она выводит гистограммы и диаграммы размаха для всех числовых признаков в DataFrame, позволяя быстро оценить распределения и обнаружить выбросы.

Функция get_eda_plots принимает следующие аргументы:

- `df: pd.DataFrame` - исходный DataFrame, для которого нужно построить графики EDA.

- `name=None` - необязательный аргумент, позволяющий указать название датафрейма для отображения на графиках.

- `sample_frac=None` - необязательный аргумент, определяющий долю данных, которые нужно использовать для визуализации. Это может быть полезно при работе с большими наборами данных.

- `seed=SEED` - необязательный аргумент, определяющий начальное значение для генератора случайных чисел. Это используется при сэмплировании данных.

- `outliers=True` - необязательный аргумент, указывающий, нужно ли отображать информацию о выбросах на гистограммах.

- `iqr_coeff=1.5` - необязательный аргумент, определяющий коэффициент для расчета межквартильного размаха при определении выбросов.

Сначала функция проверяет необходимость сэмплирования данных и отображает гистограммы всех числовых признаков. Затем, если указано, она выводит информацию о выбросах на гистограммах. После этого функция выводит диаграммы размаха для всех числовых признаков для более подробного анализа выбросов и распределений данных.

````python
# Функция вывода графиков ИАД.
def get_eda_plots(df : pd.DataFrame, name = None,
                  sample_frac = None, seed = SEED,
                  outliers = True, iqr_coeff = 1.5):

    # Проверим необходимость сэмплирования.
    if sample_frac:
        df_to_plot = df.sample(frac=sample_frac, random_state=seed)
        sample_t = f' {sample_frac:.0%} выборки'
    else:
        df_to_plot = df
        sample_t = ''
    # Проверим указание названия датафрейма.
    name = f' датафрейма "{name}"' if name else ''

    # Выведем гистограммы.
    axs = df_to_plot.hist(figsize=[20, 20], bins=50, edgecolor='black')
    if outliers:
        for ax, col in zip(axs.flatten(), df_to_plot.columns):
            outlier_perc = get_outliers_ratio(df[col], iqr_coeff=iqr_coeff)
            ax.set_title(f'Признак "{col}".\nВыбросов: '+
                         f'{outlier_perc:.1%}')
    plt.suptitle('Гистограммы' + sample_t + name + '.', y=.99, fontsize=16)
    plt.tight_layout()
    plt.show();

    # Выведем диаграммы размаха.
    df_to_plot.plot(kind='box', grid=True, figsize=[20, 8])
    plt.title('Диаграммы размаха'  + sample_t + name + '.', fontsize=16)
    plt.tight_layout()
    plt.show();
    
hide_toggle()
````

### Поиск оптимального количчества кластеров методом локтя

````python
#Inertia is within cluster sum of squares
k_list=[5, 10, 25, 50, 100, 200, 400, 600, 1000]
def get_inertia():
    inertia = []
    for n_clusters in tqdm(k_list):
        model = KMeans(n_clusters=n_clusters, random_state=SEED, init="k-means++")
        model.fit(X_base_scaled_df)
        inertia.append(model.inertia_)

    plt.figure(figsize = (20, 6))
    plt.plot(list(k_list), inertia , 'o')
    plt.plot(list(k_list), inertia , '-' , alpha = 0.25)
    plt.xlabel('Number of Clusters') , plt.ylabel('Inertia')
    plt.show()
````

### Функция алгоритма поиска похожих KMeans + NearestNeighbors

````python
def get_kmeans_acc5(df: pd.DataFrame, 
                   df_eval: pd.DataFrame,
                   target: pd.Series,
                   verbose = 1):
    if verbose > 0:
        def verboseprint(*args):
            for arg in args:
                print(arg),
            print
    else:
        verboseprint = lambda *a: None
        

        
    verboseprint('Объявляем модель KMeans с параметором n_clusters=100') 
    verboseprint('~'*10)
    kmeans = KMeans(n_clusters=100, random_state=SEED, n_init="auto", init='k-means++').fit(df)
    
    verboseprint('Создадим в исходном датафрейме признак cluster и разметим его данными из kmeans.labels_')
    verboseprint('~'*10)
    df['cluster'] = kmeans.labels_
    
    verboseprint('Создадим в тренировочном датафрейме признак cluster и разметим его данными из kmeans.predict()')
    verboseprint('~'*10)
    df_eval['cluster'] = kmeans.predict(df_eval)
    
    verboseprint('Объявим мобель NearestNeighbors')
    verboseprint('~'*10)
    neigh = NearestNeighbors(n_neighbors=5, metric='euclidean')
    
    verboseprint('Объявим список для хранения соседей')
    verboseprint('~'*10)
    neighbors = []
    
    verboseprint('Начинаем поиск похожих для кажого объекта из набора данных')
    verboseprint('~'*10)
    
    df_eval_neighbors = pd.DataFrame(data=0, index=df_eval.index, 
                                 columns=['preds_1', 'preds_2', 'preds_3', 'preds_4', 'preds_5'])

    for cluster in df_eval['cluster'].unique():
        neigh.fit(df.loc[df['cluster'] == cluster])
        df_eval_cluster = df_eval.loc[df_eval['cluster'] == cluster]
        dist, indices = neigh.kneighbors(df_eval_cluster, 5)
        df_eval_neighbors.loc[df_eval['cluster'] == cluster] = indices
    
    
    verboseprint('Начинаем расчет метрики качества')
    verboseprint('~'*10)
    acc_kmeans = 0
    for t, el in zip(target.values.tolist(), df_eval_neighbors.values.tolist()):
        acc_kmeans += int(t in [base_index[r] for r in el])
    
    df.drop(['cluster'], axis=1, inplace=True)
    df_eval.drop(['cluster'], axis=1, inplace=True)
    return round((100 * acc_kmeans / len(target)), 3)
````

## FAISS

### Функция вычисления accuracy5 для FAISS

````python
def get_faiss_acc5(df: pd.DataFrame, 
                   df_eval: pd.DataFrame,
                   target: pd.Series,
                   verbose = 1):
    
    if verbose > 0:
        def verboseprint(*args):
            for arg in args:
                print(arg),
            print
    else:
        verboseprint = lambda *a: None
        
    k = 5 # объявление количества соседей для поиска

    dims = df.shape[1] # объявление размерности на основе сета данных
    verboseprint('Определяем размерность', dims)
    verboseprint('~'*10)
    
    # привести сеты данных к формату требуемому FAISS
    verboseprint('Приводим данные к формату FAISS')
    verboseprint('~'*10)
    df_arr = np.ascontiguousarray(df.values[:2500000,:]).astype('float32')
    df_eval_arr = np.ascontiguousarray(df_eval.values).astype('float32')
    
    
    # объявление индекса на основе объявленной размерности
    verboseprint('Объявляем индекс')
    verboseprint('~'*10)
    
    faiss_n_clust = 45
    quantizer = faiss.IndexFlatL2(dims)
    index = faiss.IndexIVFFlat(quantizer,dims,faiss_n_clust)
    index.train(df_arr)
    
    # проверка обучения и добавление данных в индекс
    verboseprint('Индекс обучен:')
    print(index.is_trained)
    index.add(df_arr)
    verboseprint('~'*10)
    verboseprint('index.ntotal:')
    print(index.ntotal)
    verboseprint('~'*10)
    
    verboseprint('Проверка работоспособности.')
    verboseprint('Первые 5 объектов похожи сами на себя')
    D, I = index.search(df_arr[:5], k) # проверка работоспособности
    print(I)
    verboseprint('~'*10)
    
    verboseprint('Старт поиска похожих на полном объём данных')
    verboseprint('~'*10)
    D, I = index.search(df_eval_arr, k)
    
    verboseprint('Старт расчета метрики')
    acc_faiss = 0
    for t, el in zip(target.values.tolist(), I):
        acc_faiss += int(t in [base_index[r] for r in el])
    
    verboseprint('~'*10)
    verboseprint('Завершение работы')
    verboseprint('~'*10)
    return round((100 * acc_faiss / len(I)), 3)
````
